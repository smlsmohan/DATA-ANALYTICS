```python
import pandas as pds

#loading Employee CSV 
column_names = ["sepal_length_in_cm","sepal_width_in_cm","petal_length_in_cm","petal_width_in_cm","class"]

data = pds.read_csv('C:/Users/HP/Downloads/Foundations of AI/iris.data',names=column_names)

#Displaying Summary
data.head()
```

```python
#Features of IRIS Dataset
list(data.columns[0:-1])
```

```python
# Target class and its unique values
print("Target class in the IRIS dataset:")
print(data['class'].unique())
```

```python
data.describe()
```

```python
# Compute min, max, standard deviation, and median
data_dispersion = data.describe().loc[['min', 'max', 'std', '50%']]
print(data_dispersion)
```

```python
# Computing data dispersion features
numeric_columns = ["sepal_length_in_cm","sepal_width_in_cm","petal_length_in_cm","petal_width_in_cm"]
    
min_values = data[numeric_columns].min()
max_values = data[numeric_columns].max()
std_values = data[numeric_columns].std()
median_values = data[numeric_columns].median()

print("Minimum values:\n", min_values)
print("Maximum values:\n", max_values)
print("Standard deviation:\n", std_values)
print("Median values:\n", median_values)
```

```python
#Histogram
import matplotlib.pyplot as plt

data.hist(figsize=(10, 6), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histogram of IRIS Dataset Features")
plt.show()
```

```python
#boxplot
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title("Boxplot of IRIS Features")
plt.show()
```

```python
#Scatter Plot
sns.pairplot(data, hue="class", markers=["o", "s", "D"])
plt.show()
```
